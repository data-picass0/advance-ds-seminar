```python
from datasets import load_dataset
```

```python
# Load SQuAD dataset from Hugging Face
squad_dataset = load_dataset("squad")
```

```python
def generate_response_instruction(input_field):
    if input_field:
        return "### Instruction:\n{instruction}\n\n### Input:\n{input}\n\n### Response:"
    else:
        return "### Instruction:\n{instruction}\n\n### Response:"
```

```python
self_instruct_dataset = []

for example in squad_dataset["train"]:
    context = example["context"]
    question = example["question"]
    answer = example["answers"]["text"][0] if len(example["answers"]["text"]) > 0 else "N/A"
    input_field = True if context else False

    instruction_template = generate_response_instruction(input_field)
    instruction = instruction_template.format(instruction=example["question"], input=context)
    
    self_instruct_example = {
        "instruction": instruction,
        "input": {"context": context, "question": question} if input_field else {},
        "output": {"answer": answer}
    }

    self_instruct_dataset.append(self_instruct_example)
```
